# 🛠️ Preparación de los Datos

## Librerias a utilizar
* Pandas 🐼
* matplotlib
* seaborn
* scikit-learn
* imbalanced-learn

```python
import pandas as pd
```

```python
df = pd.read_csv('Datos_Telecom_Tratados_1.csv')
df
```

## Eliminación columnas Irrelevantes

```python
df = df.drop(columns=['ID_Cliente', 'Cuentas_Diarias'])
df
```

## Encoding

```python
df_codificado = pd.get_dummies(df, drop_first=True, dtype=int)
```

```python
print("Dimensiones antes de codificar:", df.shape)
print("Dimensiones DESPUÉS de codificar:", df_codificado.shape)
```

```python
# Damos un vistazo a las primeras filas para ver la transformación
df_codificado.head()
```

## Verificación de la Proporción de Cancelación (Churn)

```python
# Clientes que cancelaron -> 0
# Clientes que no cancelaron -> 1
conteo = df['Evasion'].value_counts()
conteo
```

```python
# Obtenemos la proporción en porcentaje
proporcion_conteo = df['Evasion'].value_counts(normalize=True)*100
proporcion_conteo
```

## Balanceo de Clases

```python
from imblearn.over_sampling import SMOTE
```

```python
# Separamos X (Variables predictorias) y 'y' (Variable Objetivo)
X = df_codificado.drop('Evasion', axis=1)
y = df_codificado['Evasion']
```

```python
# Antes de aplicar SMOTE
# Clientes que cancelaron -> 0
# Clientes que no cancelaron -> 1
y.value_counts()
```

```python
# Usamos random_state=42 para que los datos sintéticos sean reproducibles
smote = SMOTE(random_state=42)

# Aplicamos Tecnicas de Balanceo
X_balanceado, y_balanceado = smote.fit_resample(X, y)
```

```python
# Despues de Smote
y_balanceado.value_counts()
```

```python
print(f"Nuevas dimensiones de X: {X_balanceado.shape}")
print(f"Nuevas dimensiones de y: {y_balanceado.shape}")
```

## Normalización o Estandarización

```python
from sklearn.preprocessing import StandardScaler

# Instanciamos el estandarizador
scaler = StandardScaler()
```

```python
columnas_numericas = ['Meses_Contrato', 'Cargo_Mensual', 'Cargo_Total']
```

```python
X_balanceado[columnas_numericas] = scaler.fit_transform(X_balanceado[columnas_numericas])
X_balanceado[columnas_numericas].head()
```

# 🎯 Correlación y Selección de Variables

## Análisis de Correlación

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Unimos temporalmente nuestras variables predictoras y la variable objetivo
df_analisis = X_balanceado.copy()
df_analisis['Evasion'] = y_balanceado
df_analisis
```

```python
# Calculamos la matriz de correlación
matriz_correlacion = df_analisis.corr()
matriz_correlacion
```

```python
# Extraemos y ordenamos la correlación específica con la Cancelación (Evasion)
correlacion_target = matriz_correlacion['Evasion'].sort_values(ascending=False)
correlacion_target
```

```python
# Visualizamos la matriz de correlación con un Heatmap
plt.figure(figsize=(22, 16))
sns.heatmap(matriz_correlacion, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación de Variables (Heatmap)', fontsize=18)
plt.tight_layout()
plt.show()
```

## Análisis Dirigido

```python
# Configuramos el estilo visual
sns.set_theme(style="whitegrid")
```

```python
# Creamos un "lienzo" con 2 espacios (1 fila, 2 columnas)
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Tiempo de Contrato vs Cancelación
sns.boxplot(data=df, x='Evasion', y='Meses_Contrato', hue='Evasion', palette='pastel', legend=False, ax=ax[0])
ax[0].set_title('Impacto del Tiempo de Contrato en la Cancelación', fontsize=14)
ax[0].set_xlabel('Canceló el Servicio (0 = No, 1 = Sí)', fontsize=12)
ax[0].set_ylabel('Meses de Contrato', fontsize=12)

# Gráfico 2: Gasto Total vs Cancelación
sns.boxplot(data=df, x='Evasion', y='Cargo_Total', hue='Evasion', palette='pastel', legend=False, ax=ax[1])
ax[1].set_title('Impacto del Gasto Total en la Cancelación', fontsize=14)
ax[1].set_xlabel('Canceló el Servicio (0 = No, 1 = Sí)', fontsize=12)
ax[1].set_ylabel('Cargo Total', fontsize=12)

plt.tight_layout()
plt.show()
```

# 🤖 Modelado Predictivo

## Separación de Datos

```python
from sklearn.model_selection import train_test_split
```

```python
# Dividimos los datos (80% porciento entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(
    X_balanceado,
    y_balanceado,
    test_size=0.2,
    random_state=42,
    stratify=y_balanceado
)
```

```python
print(f"Datos para Entrenar (X_train): {X_train.shape[0]} filas")
print(f"Datos para Probar (X_test): {X_test.shape[0]} filas")
```

```python
# Verificamos que se haya cumplido la regla del equilibrio
print("Equilibrio en Entrenamiento (y_train)")
print(y_train.value_counts(normalize=True) * 100)
```

```python
print("Equilibrio en Prueba (y_test)")
print(y_test.value_counts(normalize=True) * 100)
```

## Creación de Modelos

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

### Regresión Logística

```python
modelo_rl = LogisticRegression(max_iter=1000, random_state=42)
```

```python
# Entrenamos al modelo
modelo_rl.fit(X_train, y_train)

# Hacemos pruebas con los datos de prueba
predicciones_rl = modelo_rl.predict(X_test)
```

### Random Forest 🌲

```python
#n_estimators=100 crea 100 arboles de decisión diferentes
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Entrenamos al modelo
modelo_rf.fit(X_train, y_train)

predicciones_rf = modelo_rf.predict(X_test)
```

## Evaluación de los Modelos

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
import matplotlib.pyplot as plt
```

```python
# 1. Obtenemos las predicciones sobre los datos de entrenamiento (para checar overfitting)
pred_train_rl = modelo_rl.predict(X_train)
pred_train_rf = modelo_rf.predict(X_train)
```

### Regresión Logistica

```python
print(f"Exactitud en Entrenamiento: {accuracy_score(y_train, pred_train_rl):.4f}")
print(f"Exactitud en Prueba (Test): {accuracy_score(y_test, predicciones_rl):.4f}")
print(f"Reporte de Clasificación con datos de Prueba:")
print(classification_report(y_test, predicciones_rl))
```

### Random Forest 🌲

```python
print(f"Exactitud en Entrenamiento: {accuracy_score(y_train, pred_train_rf):.4f}")
print(f"Exactitud en Prueba (Test): {accuracy_score(y_test, predicciones_rf):.4f}")
print("Reporte de Clasificación (Datos de Prueba):")
print(classification_report(y_test, predicciones_rf))
```

```python
# 3. Dibujamos las Matrices de Confusión
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Matriz Regresión Lógistica
ConfusionMatrixDisplay.from_predictions(y_test, predicciones_rl, ax=axes[0], cmap='Blues', colorbar=False)
axes[0].set_title('Matriz de Confusión - Regresión Logística', fontsize=12)
axes[0].set_xlabel('Predicción del Modelo')
axes[0].set_ylabel('Valor Real')

# Matriz Random Forest
ConfusionMatrixDisplay.from_predictions(y_test, predicciones_rf, ax=axes[1], cmap='Greens', colorbar=False)
axes[1].set_title('Matriz de Confusión - Random Forest', fontsize=12)
axes[1].set_xlabel('Predicción del Modelo')
axes[1].set_ylabel('Valor Real')

plt.tight_layout()
plt.show()
```

## 1. Resumen Ejecutivo
Se desarrolló un pipeline de Machine Learning para predecir qué clientes tienen mayor probabilidad de cancelar sus servicios (Churn). El análisis se basó en un conjunto de datos históricos que fue preprocesado (codificación One-Hot, balanceo con SMOTE y estandarización) para entrenar y evaluar dos modelos de clasificación: Regresión Logística y Random Forest. Los resultados permitieron identificar no solo quién va a cancelar, sino **por qué** lo hacen.

## 2. Evaluación de los Modelos Predictivos
Se implementaron dos enfoques distintos para garantizar la robustez del análisis:

* **Regresión Logística:** Demostró ser el modelo más estable y confiable para la inferencia. Logró una exactitud del **~81%** tanto en entrenamiento como en prueba, lo que indica una excelente capacidad de generalización sin presentar *overfitting* (sobreajuste).
* **Random Forest:** Aunque alcanzó una exactitud ligeramente superior en la prueba real (**~84%**), presentó un claro escenario de *overfitting* al memorizar los datos de entrenamiento (99.99%). Sin embargo, su estructura de árboles fue invaluable para confirmar la jerarquía de las variables más importantes.

## 3. Principales Factores que Impulsan la Cancelación
Al extraer los coeficientes matemáticos de la Regresión Logística y la reducción de impureza del Random Forest, descubrimos que la decisión de cancelar está fuertemente impulsada por tres dimensiones:

### A. Estructura del Contrato (El factor más crítico)
* **Contratos "Mes a mes":** Es la variable número uno que detona la cancelación. La falta de compromiso a largo plazo facilita la salida inmediata del cliente.
* **Antigüedad (Meses_Contrato):** Actúa como el principal escudo de la empresa. A mayor tiempo en la empresa, la probabilidad de cancelación cae drásticamente, consolidando la lealtad con contratos de uno o dos años.

### B. Sensibilidad Económica
* **Cargo Mensual y Cargo Total:** El modelo Random Forest identificó que el volumen de gasto es un nodo de decisión crítico. Los clientes son altamente sensibles al precio mensual y al acumulado, lo que sugiere que evalúan constantemente la relación costo-beneficio de su plan.

### C. Fricciones Operativas y de Servicio
* **Internet de Fibra Óptica:** Sorprendentemente, poseer este servicio tiene una correlación positiva con la cancelación. Esto levanta una alerta sobre posibles problemas de estabilidad, fallas técnicas o un precio percibido como excesivo para este servicio en particular.
* **Método de Pago (Cheque Electrónico):** Los clientes que usan este método de pago manual o semi-manual tienen una mayor tasa de abandono en comparación con aquellos que tienen cargos automáticos a tarjetas o transferencias.

---

## 4. Estrategias de Retención Propuestas
Basados en la evidencia de los datos, proponemos a la gerencia las siguientes acciones inmediatas:

1. **Campaña de Migración de Contratos:**
   * *Acción:* Diseñar ofertas agresivas (ej. descuentos en el cargo mensual o meses gratis) dirigidas específicamente a los clientes que actualmente están en la modalidad "Mes a mes", con el objetivo de anclarlos a contratos de 1 o 2 años.
2. **Auditoría Técnica al Servicio de Fibra Óptica:**
   * *Acción:* El equipo técnico debe investigar la infraestructura de la fibra óptica en las zonas con mayor tasa de churn. Si el servicio es inestable, se debe mejorar; si la calidad es buena, el equipo de marketing debe reestructurar el precio o el valor percibido del paquete.
3. **Automatización de Pagos:**
   * *Acción:* Crear incentivos (pequeños descuentos o gigas extra) para que los clientes que pagan con "Cheque electrónico" migren a métodos de pago automáticos (tarjeta de crédito/débito). Un pago automático reduce la fricción y la decisión consciente de "tener que pagar" cada mes.
4. **Despliegue del Modelo Preventivo:**
   * *Acción:* Integrar la Regresión Logística en el CRM de la empresa para que, cada semana, genere una "Lista de Clientes en Riesgo". El equipo de atención al cliente podrá llamar a estos usuarios y ofrecerles promociones de fidelidad *antes* de que soliciten la cancelación.
