# 🛠️ Preparación de los Datos

## Librerias a utilizar
* Pandas 🐼
* matplotlib
* seaborn
* scikit-learn
* imbalanced-learn

```python
import pandas as pd
```

```python
df = pd.read_csv('Datos_Telecom_Tratados_1.csv')
df
```

## Eliminación columnas Irrelevantes

```python
df = df.drop(columns=['ID_Cliente', 'Cuentas_Diarias'])
df
```

## Encoding

```python
df_codificado = pd.get_dummies(df, drop_first=True, dtype=int)
```

```python
print("Dimensiones antes de codificar:", df.shape)
print("Dimensiones DESPUÉS de codificar:", df_codificado.shape)
```

```python
# Damos un vistazo a las primeras filas para ver la transformación
df_codificado.head()
```

## Verificación de la Proporción de Cancelación (Churn)

```python
# Clientes que cancelaron -> 0
# Clientes que no cancelaron -> 1
conteo = df['Evasion'].value_counts()
conteo
```

```python
# Obtenemos la proporción en porcentaje
proporcion_conteo = df['Evasion'].value_counts(normalize=True)*100
proporcion_conteo
```

## Balanceo de Clases

```python
from imblearn.over_sampling import SMOTE
```

```python
# Separamos X (Variables predictorias) y 'y' (Variable Objetivo)
X = df_codificado.drop('Evasion', axis=1)
y = df_codificado['Evasion']
```

```python
# Antes de aplicar SMOTE
# Clientes que cancelaron -> 0
# Clientes que no cancelaron -> 1
y.value_counts()
```

```python
# Usamos random_state=42 para que los datos sintéticos sean reproducibles
smote = SMOTE(random_state=42)

# Aplicamos Tecnicas de Balanceo
X_balanceado, y_balanceado = smote.fit_resample(X, y)
```

```python
# Despues de Smote
y_balanceado.value_counts()
```

```python
print(f"Nuevas dimensiones de X: {X_balanceado.shape}")
print(f"Nuevas dimensiones de y: {y_balanceado.shape}")
```

## Normalización o Estandarización

```python
from sklearn.preprocessing import StandardScaler

# Instanciamos el estandarizador
scaler = StandardScaler()
```

```python
columnas_numericas = ['Meses_Contrato', 'Cargo_Mensual', 'Cargo_Total']
```

```python
X_balanceado[columnas_numericas] = scaler.fit_transform(X_balanceado[columnas_numericas])
X_balanceado[columnas_numericas].head()
```

# 🎯 Correlación y Selección de Variables

## Análisis de Correlación

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Unimos temporalmente nuestras variables predictoras y la variable objetivo
df_analisis = X_balanceado.copy()
df_analisis['Evasion'] = y_balanceado
df_analisis
```

```python
# Calculamos la matriz de correlación
matriz_correlacion = df_analisis.corr()
matriz_correlacion
```

```python
# Extraemos y ordenamos la correlación específica con la Cancelación (Evasion)
correlacion_target = matriz_correlacion['Evasion'].sort_values(ascending=False)
correlacion_target
```

```python
# Visualizamos la matriz de correlación con un Heatmap
plt.figure(figsize=(22, 16))
sns.heatmap(matriz_correlacion, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación de Variables (Heatmap)', fontsize=18)
plt.tight_layout()
plt.show()
```

## Análisis Dirigido

# 🤖 Modelado Predictivo

## Separación de Datos

```python
from sklearn.model_selection import train_test_split
```

```python
# Dividimos los datos (80% porciento entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(
    X_balanceado,
    y_balanceado,
    test_size=0.2,
    random_state=42,
    stratify=y_balanceado
)
```

```python
print(f"Datos para Entrenar (X_train): {X_train.shape[0]} filas")
print(f"Datos para Probar (X_test): {X_test.shape[0]} filas")
```

```python
# Verificamos que se haya cumplido la regla del equilibrio
print("Equilibrio en Entrenamiento (y_train)")
print(y_train.value_counts(normalize=True) * 100)
```

```python
print("Equilibrio en Prueba (y_test)")
print(y_test.value_counts(normalize=True) * 100)
```

## Creación de Modelos

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

### Regresión Logística

```python
modelo_rl = LogisticRegression(max_iter=1000, random_state=42)
```

```python
# Entrenamos al modelo
modelo_rl.fit(X_train, y_train)

# Hacemos pruebas con los datos de prueba
predicciones_rl = modelo_rl.predict(X_test)
```

### Random Forest 🌲

```python
#n_estimators=100 crea 100 arboles de decisión diferentes
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Entrenamos al modelo
modelo_rf.fit(X_train, y_train)

predicciones_rf = modelo_rf.predict(X_test)
```

## Evaluación de los Modelos

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
import matplotlib.pyplot as plt
```

```python
# 1. Obtenemos las predicciones sobre los datos de entrenamiento (para checar overfitting)
pred_train_rl = modelo_rl.predict(X_train)
pred_train_rf = modelo_rf.predict(X_train)
```

### Regresión Logistica

```python
print(f"Exactitud en Entrenamiento: {accuracy_score(y_train, pred_train_rl):.4f}")
print(f"Exactitud en Prueba (Test): {accuracy_score(y_test, predicciones_rl):.4f}")
print(f"Reporte de Clasificación con datos de Prueba:")
print(classification_report(y_test, predicciones_rl))
```

### Random Forest 🌲

```python
print(f"Exactitud en Entrenamiento: {accuracy_score(y_train, pred_train_rf):.4f}")
print(f"Exactitud en Prueba (Test): {accuracy_score(y_test, predicciones_rf):.4f}")
print("Reporte de Clasificación (Datos de Prueba):")
print(classification_report(y_test, predicciones_rf))
```

```python
# 3. Dibujamos las Matrices de Confusión
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Matriz Regresión Lógistica
ConfusionMatrixDisplay.from_predictions(y_test, predicciones_rl, ax=axes[0], cmap='Blues', colorbar=False)
axes[0].set_title('Matriz de Confusión - Regresión Logística', fontsize=12)
axes[0].set_xlabel('Predicción del Modelo')
axes[0].set_ylabel('Valor Real')

# Matriz Random Forest
ConfusionMatrixDisplay.from_predictions(y_test, predicciones_rf, ax=axes[1], cmap='Greens', colorbar=False)
axes[1].set_title('Matriz de Confusión - Random Forest', fontsize=12)
axes[1].set_xlabel('Predicción del Modelo')
axes[1].set_ylabel('Valor Real')

plt.tight_layout()
plt.show()
```
